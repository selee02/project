```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.alert import Alert
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
```

```python
df2 = pd.read_csv('../코스피200ticker.csv', dtype={'ticker':str})
df2
```

```python


# df1: 종목코드가 기준 (데이터가 더 적음)
# df2: ticker가 기준 (데이터가 더 많음)

# 병합 실행
merged_df = pd.merge(
    df1, 
    df2, 
    left_on='종목코드',  # df1의 기준 컬럼명
    right_on='ticker',   # df2의 기준 컬럼명
    how='left'           # df1(왼쪽)에 있는 것만 남기기
)

# 만약 '종목코드'와 'ticker'가 같은 값이니 컬럼이 중복되는 게 싫다면 하나를 지워줍니다.
merged_df = merged_df.drop('ticker', axis=1)

merged_df
```

```python
def news_download(search_text : str, begin : str, end : str) -> None:
    
    search_element = driver.find_element(By.ID, "total-search-key")
    search_element.send_keys(search_text)
    detail_search_button = driver.find_element(By.ID, 'ig-sd-btn')
    detail_search_button.click()
    time.sleep(1)

    start_date = driver.find_element(By.ID, "search-begin-date")
    end_date = driver.find_element(By.ID, "search-end-date") 
    start_date.click()
    start_date.send_keys(Keys.CONTROL, "a")
    start_date.send_keys(Keys.DELETE)
    start_date.send_keys(begin)
    time.sleep(1)

    end_date.click()
    end_date.send_keys(Keys.CONTROL, "a")
    end_date.send_keys(Keys.DELETE)
    end_date.send_keys(end)
    time.sleep(1)

    detail_tab = driver.find_element(By.XPATH, "//a[contains(@class, 'tab-btn') and text()='상세검색']")
    detail_tab.click()
    time.sleep(1)

    scope_select = Select(driver.find_element(By.ID, "search-scope-type"))
    scope_select.select_by_value("2")
    time.sleep(1)

    search_btn = driver.find_element(By.CSS_SELECTOR, ".apply-btn.primary-btn.news-search-btn")
    search_btn.click()
    time.sleep(3)
    loader = driver.find_element(By.CSS_SELECTOR, ".news-loader.loading")

    WebDriverWait(driver, 120).until(
        lambda d: loader.get_attribute("style") == "display: none;"
    )
    filter_tab = driver.find_element(By.ID, 'filterTab03')
    filter_check = filter_tab.find_elements(By.TAG_NAME, 'span')[1]
    filter_check.click()
    time.sleep(1)

    step3_btn = driver.find_element(By.ID, "collapse-step-3")
    step3_btn.click()
    time.sleep(3)
    down_btn = driver.find_element(By.CLASS_NAME, 'news-download-btn')
    down_btn.click()
    time.sleep(1)
    try:
        alert = Alert(driver)
        alert.accept()
    except:
        pass
    time.sleep(1)
    
    loader = driver.find_element(By.CSS_SELECTOR, ".news-loader.loading")

    WebDriverWait(driver, 120).until(
        lambda d: loader.get_attribute("style") == "display: none;"
    )
    print(f"{search_text}의 뉴스의 기간 {begin} ~ {end} 다운로드되었습니다.")
```
