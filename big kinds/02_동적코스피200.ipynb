```python
kospi200 = pd.read_csv('../코스피200ticker.csv', dtype={'ticker':str})
```

```python
kospi200
```

```python
ticker200 = kospi200['ticker'].unique()
```

```python
ticker24 = kospi_24['종목코드'].unique()
```

```python
nonticker = []

for ticker in ticker200:
    if ticker not in ticker24:
        nonticker.append(ticker)
```

```python
noncompany = dict()

for ticker in nonticker:
    name = stock.get_market_ticker_name(ticker)
    noncompany[name] = ticker
```

```python
df2 = pd.DataFrame(list(noncompany.items()), columns=['종목명', '종목코드'])
```
